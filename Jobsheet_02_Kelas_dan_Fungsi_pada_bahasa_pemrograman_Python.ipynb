Praktikum 01. Fungsi pada  Python

```python
# 1. Fungsi Built-in
print("Ini adalah contoh fungsi built-in")

kata = "Pemrograman"
panjang_kata = len(kata)
print(f"Panjang kata '{kata}' adalah: {panjang_kata}")

angka = [10, 5, 30, 40, 25]
nilai_maks = max(angka)
print(f"Nilai maksimum dalam daftar {angka} adalah: {nilai_maks}")

# 2. Fungsi Buatan Sendiri (User-Defined)
def kuadrat(x):
    return x ** 2
print(f"Kuadrat dari 5 adalah: {kuadrat(5)}")

def luas_persegi_panjang(p, l):
    return p * l
print(f"Luas persegi panjang: {luas_persegi_panjang(10, 5)}")

def info(nama, umur, ipk):
    print(f"Nama: {nama}, Umur: {umur}, IPK: {ipk}")
info("Budi", 22, 3.8)

def sapa(nama):
    print(f"Halo, {nama}! Selamat datang di dunia Python.")
sapa("Andi")

def keliling_persegi(s):
    return 4 * s
print(f"Keliling persegi dengan sisi 5 adalah: {keliling_persegi(5)}")
```

Praktikum 03: Kelas dan  Objek

```python
# KELAS BUKU
class Buku:
    def __init__(self, judul, pengarang, tahun_terbit):
        self.judul = judul
        self.pengarang = pengarang
        self.tahun_terbit = tahun_terbit
        self.status = "Tersedia"

    def info(self):
        print(f"Judul: {self.judul}")
        print(f"Pengarang: {self.pengarang}")
        print(f"Tahun Terbit: {self.tahun_terbit}")
        print(f"Status: {self.status}")

    def pinjam(self):
        self.status = "Dipinjam"
        print(f"Buku '{self.judul}' telah dipinjam.")

    def kembalikan(self):
        self.status = "Tersedia"
        print(f"Buku '{self.judul}' telah dikembalikan.")

buku1 = Buku("Pemrograman Python", "John Doe", 2021)
buku1.info()

buku2 = Buku("Data Science untuk Pemula", "Jane Doe", 2020)
buku2.pinjam()
buku2.kembalikan()

buku1.pinjam()
buku1.info()
print("-" * 30)

# KELAS MAHASISWA
class Mahasiswa:
    def __init__(self, nama, nim, umur):
        self.nama = nama
        self.nim = nim
        self.umur = umur

    def info(self):
        print(f"Nama: {self.nama}")
        print(f"NIM: {self.nim}")
        print(f"Umur: {self.umur} tahun\n")

    def ubah_umur(self, umur_baru):
        self.umur = umur_baru

mhs1 = Mahasiswa("Andi", "12345", 20)
mhs1.info()

mhs2 = Mahasiswa("Budi", "67890", 22)
mhs2.info()

print("Setelah mengubah umur mahasiswa1:")
mhs1.ubah_umur(21)
mhs1.info()
```

Praktikum 04: membuat dan  menggunakan atribut dan  fungsi di dalam kelas

```python
class Mobil:
    def __init__(self, merk, warna, tahun, harga):
        self.merk = merk
        self.warna = warna
        self.tahun = tahun
        self.harga = harga

    def info(self):
        print(f"Mobil {self.merk} berwarna {self.warna}, tahun {self.tahun}, harga: Rp {self.harga}")

    def hitung_diskon(self, persen):
        harga_diskon = self.harga - (self.harga * persen / 100)
        print(f"Harga setelah diskon {persen}%: Rp {harga_diskon}")

    def usia_mobil(self, tahun_sekarang):
        usia = tahun_sekarang - self.tahun
        return usia

    def perbarui_data(self, harga_baru, tahun_baru):
        self.harga = harga_baru
        self.tahun = tahun_baru
        print(f"Harga dan tahun mobil {self.merk} diperbarui menjadi Rp {self.harga} dan tahun {self.tahun}")

# Eksekusi Program
mobil1 = Mobil("Toyota", "Hitam", 2015, 300000000)
mobil2 = Mobil("Honda", "Merah", 2018, 250000000)

mobil1.info()
mobil2.info()

mobil1.hitung_diskon(10)
mobil2.hitung_diskon(15)

print(f"Usia mobil1 pada tahun 2025: {mobil1.usia_mobil(2025)} tahun")
print(f"Usia mobil2 pada tahun 2025: {mobil2.usia_mobil(2025)} tahun")

mobil1.perbarui_data(280000000, 2022)
mobil2.perbarui_data(240000000, 2021)
```

```python
# tugas pbo 2: penjualan hp (class dan modularisasi)

#class untuk hp
class Handphone:
    def __init__(self, merk, tipe, harga, stok):
        self.merk = merk
        self.tipe = tipe
        self.harga = harga
        self.stok = stok

    def cek_info(self):
        print(f"HP {self.merk} {self.tipe} | Harga: Rp {self.harga} | Stok: {self.stok}")

    def kurangi_stok(self, jumlah):
        self.stok -= jumlah

# fungsi terpisah untuk proses beli (modularisasi)
def beli_hp(hp, jumlah_beli):
    print(f"\nProses beli {jumlah_beli} unit {hp.merk} {hp.tipe}...")

    # cek stok cukup atau nggak
    if hp.stok >= jumlah_beli:
        hp.kurangi_stok(jumlah_beli)
        total_harga = hp.harga * jumlah_beli
        print(f"Transaksi berhasil. Total bayar: Rp {total_harga}")
    else:
        print(f"Transaksi gagal, stok cuma sisa {hp.stok}.")

# bagian utama program
print("Daftar HP di toko:")
hp_pertama = Handphone("Samsung", "Galaxy A54", 6000000, 5)
hp_kedua = Handphone("Xiaomi", "Poco X5", 3500000, 2)

hp_pertama.cek_info()
hp_kedua.cek_info()

# simulasi transaksi
beli_hp(hp_pertama, 2)
beli_hp(hp_kedua, 4)

print("\nSisa stok sekarang:")
hp_pertama.cek_info()
hp_kedua.cek_info()
```
