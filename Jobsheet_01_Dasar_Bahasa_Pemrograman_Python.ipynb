1. Tipe Data Dasar

```python
angka_int = 10
print(f"angka_int= {angka_int} -> {type(angka_int)}")

angka_float = 3.14
print(f"angka_float= {angka_float} -> {type(angka_float)}")

angka_complex = 2+3j
print(f"angka_complex= {angka_complex} -> {type(angka_complex)}")

is_active = True
print(f"is_active = {is_active} -> {type(is_active)}")

teks = "Hello, Python!"
print(f"teks= {teks} -> {type(teks)}")

daftar_angka = [1, 2, 3, 4, 5]
print(f"daftar_angka= {daftar_angka} -> {type(daftar_angka)}")

koordinat = (10, 20)
print(f"koordinat = {koordinat} -> {type(koordinat)}")

data_mahasiswa = {'nama': 'Andi', 'nim': 'A11.2022.12345', 'jurusan': 'Teknik Informatika'}
print(f"data_mahasiswa = {data_mahasiswa} -> {type(data_mahasiswa)}")

himpunan_angka = {1, 2, 3}
print(f"himpunan_angka= {himpunan_angka} -> {type(himpunan_angka)}\n")

nilai_str = "100"
print(f"nilai_str = {nilai_str} -> {type(nilai_str)}")

nilai_int = int(nilai_str)
print(f"nilai_int= {nilai_int} -> {type(nilai_int)}")
```

2. Variable dan operasi dasar

```python
nama = "Budi"
umur = 20
tinggi = 170.5
mahasiswa = True

print(f"Nama {nama}")
print(f"Umur = {umur}")
print(f"Tinggi = {tinggi} cm")
print(f"Mahasiswa = {mahasiswa}\n")

print("OPERASI ARITMETIKA")
a = 10
b = 3
print(f"a= {a} , b = {b}")
print(f"Penjumlahan = {a + b}")
print(f"Pengurangan= {a - b}")
print(f"Perkalian = {a * b}")
print(f"Pembagian = {a / b}")
print(f"Pembagian Bulat= {a // b}")
print(f"Modulus = {a % b}")
print(f"Pangkat = {a ** b}\n")

print("OPERASI PERBANDINGAN")
print(f"a > b = {a > b}")
print(f"a < b = {a < b}")
print(f"a == b = {a == b}")
print(f"a != b = {a != b}")
print(f"a >= b = {a >= b}")
print(f"a <= b = {a <= b}\n")

print("OPERASI LOGIKA")
x = True
y = False
print(f"x = {x} , y = {y}")
print(f"x and y = {x and y}")
print(f"x or y = {x or y}")
print(f"not x = {not x}\n")

if a > b and b > 0:
    print("Kondisi terpenuhi: a lebih besar dari b, dan b masih positif.")
```

3. Percabangan

```python
print("Contoh IF Sederhana:")
nilai = 85
if nilai > 80:
    print("Selamat! Anda lulus dengan nilai tinggi.\n")

print("Contoh IF-ELSE:")
umur = 16
if umur >= 17:
    print("Anda boleh membuat SIM.")
else:
    print("Anda belum cukup umur untuk mendapatkan SIM.\n")

print("Contoh IF-ELIF-ELSE:")
hari = "Rabu"
if hari == "Senin":
    print("Hari Senin harga normal.")
elif hari == "Rabu":
    print("Hari Rabu Ada diskon di beberapa toko.")
else:
    print("Hari lainnya.\n")

print("Contoh IF Bersarang (Nested IF):")
suhu = 32
if suhu > 30:
    print("Cuaca cukup panas.")
    if suhu < 35:
        print("Masih relatif normal, tapi tetap jaga kesehatan.\n")

print("Contoh IF dengan Operasi Logika AND/OR:")
nilai_teori = 80
nilai_praktik = 75
if nilai_teori >= 70 and nilai_praktik >= 70:
    print("Anda lulus karena nilai teori dan praktik memadai.\n")

print("Contoh If Ternary (Conditional Expression):")
angka = -5
hasil = "Positif" if angka > 0 else "Negatif atau Nol"
print(f"Angka {angka} => {hasil}")
```

 Program yang memeriksa apakah suatu bilangan genap, ganjil,  atau prima.

```python
angka = int(input("Masukkan bilangan: "))

if angka % 2 == 0:
    jenis = "Genap"
else:
    jenis = "Ganjil"

prima = True
if angka <= 1:
    prima = False
else:
    for i in range(2, angka):
        if angka % i == 0:
            prima = False
            break

if prima:
    status = "Prima"
else:
    status = "Bukan Prima"

print(f"Angka {angka} adalah bilangan {jenis} dan {status}")
```
